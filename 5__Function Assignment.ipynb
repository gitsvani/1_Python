```python
def OddEven():
    num = int(input('Input any number : '))
        
    if ((num % 2) == 0):
        print(num, "is an even number")
    else :
        print(num, "is an odd number")
```

```python
def Elegible():
    gender = input("Input your gender : ")
    age = int(input("Input your age : ")) 

    if (gender == 'Male' or gender == 'male'):
        if (age < 21):
            print("You are not eligible for marriage")
        else:
            print("You are eligible for marriage")
            
    elif (gender == 'Female' or gender == 'female'):
        if (age < 18):
            print("You are not eligible for marriage")
        else:
            print("You are eligible for marriage")      
    else:
        print("Gender input error")
        
```

```python
Elegible() 
```

```python
def percentage():
    
    num = int(input("Enter the total number of subjects : "))

    score = 0.0
    total = 0.0
    for cyc in range(1,num+1):
        score = float(input(f"Enter the score for Subject{cyc}: "))
        total = total + score

    percent = total / num
    print (f"Your total score : {total}")
    print (f"Percentage : {percent}"+"%")         # I want to know how to convert variable to float to get answer as 29.59999999
```

```python
percentage()
```

```python

# Area of a triangle #
height1 = float(input("Enter the height of the Triangle-1 : "))
breadth1 = float(input("Enter the breadth of the Triangle-1 : "))

area = (height1 * breadth1) / 2

print ("Formula to calculate area of a triangle  : (Height * Breadth) / 2")
print (f"Area of Triangle  : {area}")
print ()

# Perimeter of a triangle #
peri_hgt1 = int(input("Enter height-1 of the triangle : "))
peri_hgt2 = int(input("Enter height-2 of the triangle : "))
peri_brt = int(input("Enter breadth of the triangle : "))

perimtr = peri_hgt1 + peri_hgt2 + peri_brt
print ("Formula to calculate perimeter of a triangle  : Height1 + Height2 + Breadth ")
print (f"Perimeter of Triangle  : {perimtr}")
```
