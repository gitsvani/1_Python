```python
age = int( input("Enter the age:"))

if (age <= 18):
    print("Children")
elif (age > 18 and age <= 30):
    print("Adult")
elif (age > 30 and age <= 60):
    print("Citizen")  
else:
    print("Senior Citizen")
    
```
